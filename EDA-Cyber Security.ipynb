### Bivariate Analysis

```python
pd.crosstab(index=df['Attack_type'],columns=df['protocol_type']).plot(kind='bar')
plt.title("Attack_type vs Protocal_type",c='r')
plt.xticks(rotation=45)
plt.show()
```

#### Protocol 0.0 is most susceptible to Neptune Attacks.

```python
q2=df.groupby(by=['Attack_type']).agg(Bytes_transfered=('dst_bytes',sum))
```

```python
q2.drop(labels='Normal').plot(kind='bar')
plt.title("Bytes Transaferred by type of attacks",c='r')
plt.ylabel("Bytes")
plt.show()
```

#### Unknown attack type tend to transfer more data followed by PortSweep adn BufferOverflow.

```python
q3=df.loc[(df['attack_flag']==1)]
```

```python
q3.groupby(by=['Attack_type']).agg(Failed_Login_Attempt=('num_failed_logins','count')).plot(kind='bar')
plt.title("Attack_type vs Failed login attempts",c='r')
plt.show()
```

#### Neptune has most number of failed login attempts.
